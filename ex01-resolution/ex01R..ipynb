## 1. Define:

### (a) Subgraph

R: Considerando um grafo G(V, A), onde V é o conjunto de vértices e A o conjunto de arestas, o Grafo G' é um subgrafo de G, acaso G'(V', A') com V' sendo o conjunto de vértices do grafo G' e A' o conjunto de arestas de G' e tanto G' como A' estão contidos em V e A, respectivamente.

### (b) Bipartite graph

R: Considerando um grafo G(V, A), um grafo bipartido é um grafo onde é possível dividir o conjunto de vértices V em dois outros cunjuntos (V' e V'') onde não há qualquer aresta ligando os vértices de um mesmo conjunto mas apenas com vértices do outro conjunto.

### (c) Hamiltonian graph

R: Um grafo hamiltoniano é um grafo que contém um clico hamiltoniano que é um ciclo onde é possível percorrer um caminho saindo de vértice e visitando todos os vértices do grafo apenas uma vez e voltando para o vértice inicial.

### (d) Eulerian graph.

R: Um grafo euliriano é um grafo que contém um ciclo euliriano que é o tipo de ciclo onde é possível, saindo de um vértice inicial, visitar todas as arestas presentes no grafo e depois retornar ao vértice inicial.

## 2. Describe how a breadth-first search algorithm works.

R: O algoritmo de busca em largura consiste em um algoritmo recursivo onde o objeto é buscar um vértice desejado em um grafo. Utilizando uma fila, escolhe-se um vértice aleatoriamente e marca esse vértice colocando-o na fila junto com seus vértices adjacentes. Depois o proximo passo é ir no primeiro vértice da fila, retira-lo da fila e colocar os vértices adjacentes à esse vértice ao final da fila até que o vertice desejado seja encontrado ou todos os vértices do grafo seja marcado.

## 3. How many edges does a complete graph with n vertices have? What about a complete directed graph with n vertices?


R: Um grafo completo é chamado de Kn onde n é o número de vértices desse grafo. Considerando que cada vértice irá ser adjacente (ter uma aresta ligando) à todo outro vértice do grafo, deve considerar a ligação de todos com o n vértices n-1 vezes, pois não irá existir aresta saindo de um vértice e indo para ele mesmo e também dividir por dois, pois não deve-se considerar uma aresta na ida e na volta no caso de um grafo não direcionado. Sendo assim através da fórmula: n(n-1)/2 pode-se encontrar a quantidade de arestas de um grafo completo (Kn).

## 8. How many edges does a graph have 3 vertices of degree 3 and one vertex of degree 5?

R: Através da fórmula: 2E= ∑ v∈V * d(v), com E sendo o número de arestas do grafo, V o conjunto de vértices e d(v) o grau do vértice. Ao ter as informações de que existem três vértices de grau 3 e um vértice de grau 5, fazendo as substituições:
2E = 5 + 3 + 3 + 3
2E = 14
E = 14/2
E = 7
Sendo assim 7 é o número de arestas do grafo.

### 10. Draw the graph from the previous exercise as undirected and weighted, with the weight being 2 if the connection is reciprocal, 1 otherwise.

```python

import matplotlib.pyplot as plt

G = nx.Graph()

G.add_edge('A', 'B', weight=1)
G.add_edge('A', 'G', weight=2)
G.add_edge('B', 'C', weight=2)
G.add_edge('B', 'D', weight=2)
G.add_edge('C', 'D', weight=1)
G.add_edge('D', 'G', weight=1)
G.add_node('E')
G.add_node('F')
G.add_node('H')

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)

labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

# Exibir o gráfico
plt.show()
```
